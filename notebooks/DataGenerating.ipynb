# Generate Data for Arabic & Spanish speakers

```python
# Define paths
original_data_path = '/content/drive/My Drive/Deep_Project/Data'  # Path in Google Drive
new_data_path = "/content/drive/My Drive/Deep_Project/new_data"
target_size_mb = 256  # Target total size in MB
num_speakers = 8  # Number of speakers to include

# Create or clean the destination folder
if os.path.exists(new_data_path):
  #print contents
  print(f"Contents of {new_data_path}:")
  for filename in os.listdir(new_data_path):
    print(filename)
  # delete contents of folder
  for filename in os.listdir(new_data_path):
    shutil.rmtree(os.path.join(new_data_path, filename))
    print(f"Deleted {filename}")
else:
  print(f"Folder {new_data_path} does not exist")
```

```python
# Calculate the size limit per speaker
size_per_speaker = float(target_size_mb) / num_speakers
total_speakers=0
chosen_speakers=0
# List speaker folders
speakers = [folder for folder in os.listdir(original_data_path) if os.path.isdir(os.path.join(original_data_path, folder))]
for speaker in speakers:
  total_speakers += 1
speakers = sorted(speakers)[:total_speakers]  # Select the first speakers
# print numer of entries in speakers, and then print speakers
print(f"Number of speakers: {len(speakers)}")
print(f"Speakers: {speakers}")

# Process each speaker
for idx, speaker in enumerate(speakers):
    if (speaker!=speakers[0] and speaker!=speakers[1] and speaker!=speakers[2] and speaker!=speakers[3] and speaker!=speakers[16] and speaker!=speakers[17] and speaker!=speakers[18] and speaker!=speakers[19]):
      continue
    chosen_speakers += 1
    print(f"Processing speaker folder: {speaker}")
    speaker_path = os.path.join(original_data_path, speaker, 'wav')

    # Check if the rawaudio folder exists
    if not os.path.exists(speaker_path):
        print(f"Skipping {speaker}: 'wav' folder not found.")
        continue

    # Get all audio files for the speaker
    audio_files = [f for f in os.listdir(speaker_path) if os.path.isfile(os.path.join(speaker_path, f))]
    if not audio_files:
        print(f"No audio files found for {speaker}. Skipping.")
        continue

    random.shuffle(audio_files)  # Shuffle for randomness

    # Size limit in bytes
    size_limit_bytes = size_per_speaker * 1024 * 1024
    selected_files = []
    total_size = 0

    # Select files within the size limit
    for file in audio_files:
        file_path = os.path.join(speaker_path, file)
        try:
            file_size = os.path.getsize(file_path)
        except Exception as e:
            print(f"Error getting size of file {file}: {e}")
            continue

        if total_size + file_size > size_limit_bytes:
            break

        selected_files.append(file)
        total_size += file_size

    # Create new speaker folder
    new_speaker_folder = os.path.join(new_data_path, f'speaker{idx + 1}')
    os.makedirs(new_speaker_folder, exist_ok=True)

    # Copy selected files
    for file in selected_files:
        try:
            shutil.copy(os.path.join(speaker_path, file), os.path.join(new_speaker_folder, file))
        except Exception as e:
            print(f"Error copying file {file}: {e}")

    print(f"Processed {speaker} -> speaker{idx + 1}, {len(selected_files)} files, {total_size / (1024 * 1024):.2f} MB")

print("Number of chosen speakers from first dataset:", chosen_speakers)
print("Dataset reduced and reorganized successfully!")
```

```python
import os
import librosa
import numpy as np
import soundfile as sf
import shutil
import random
import pyrubberband as pyrb  # For VTLP

# Augmentation functions
def add_white_noise(audio, noise_level_range=(0.002, 0.004)):
    """Add white noise with a random intensity."""
    noise_level = np.random.uniform(*noise_level_range)
    noise = np.random.randn(len(audio)) * noise_level
    return audio + noise

def pitch_shift(audio, sr, n_steps_range=(-5, 5), mean=0, std=2):
    """Shift pitch by a random number of semitones, biased toward smaller shifts."""
    # Generate a random semitone shift from a normal distribution
    n_steps = int(np.clip(np.random.normal(loc=mean, scale=std), *n_steps_range))
    return librosa.effects.pitch_shift(audio, sr=sr, n_steps=n_steps)

""" Changing pitch shift ^
def pitch_shift(audio, sr, n_steps_range=(-5, 5)):
    #Shift pitch by a random number of semitones within the range.
    n_steps = np.random.randint(*n_steps_range)
    return librosa.effects.pitch_shift(audio, sr=sr, n_steps=n_steps)
"""

def custom_time_stretch(audio, rate_range=(0.9, 1.1)):
    """Time stretch the audio by a random rate within the range."""
    rate = np.random.uniform(*rate_range)
    return librosa.effects.time_stretch(audio, rate=rate)

def vtlp(audio, sr, alpha_range=(0.9, 1.1)):
    """Apply Vocal Tract Length Perturbation (VTLP)."""
    alpha = np.random.uniform(*alpha_range)
    return pyrb.pitch_shift(audio, sr, alpha)

# Apply augmentations with probabilities
def apply_augmentations(input_folder, output_folder, combined_folder, augmentations, probabilities):
    os.makedirs(output_folder, exist_ok=True)
    os.makedirs(combined_folder, exist_ok=True)

    for file_name in os.listdir(input_folder):
        if file_name.endswith('.wav'):  # Only process .wav files
            file_path = os.path.join(input_folder, file_name)

            # Load audio file
            audio, sr = librosa.load(file_path, sr=16000)

            # Apply augmentations based on probabilities
            augmented_audio = audio  # Start with the original audio
            applied_augmentations = []  # Track applied augmentations for naming

            for aug_name, aug_func in augmentations.items():
                if random.random() < probabilities.get(aug_name, 0):  # Check probability
                    if aug_name == "background_noise":
                        augmented_audio = aug_func(augmented_audio, sr)
                    else:
                        augmented_audio = aug_func(augmented_audio, sr) if "sr" in aug_func.__code__.co_varnames else aug_func(augmented_audio)
                    applied_augmentations.append(aug_name)

            # Skip saving if no augmentations were applied
            if not applied_augmentations:
                continue

            # Generate unique augmented file name
            base_name = file_name.replace('.wav', f"_{'_'.join(applied_augmentations)}.wav")
            unique_name = f"{base_name}"

            # Save in speaker-specific folder
            output_path = os.path.join(output_folder, unique_name)
            sf.write(output_path, augmented_audio, sr)
            print(f"Saved: {output_path}")

            # Copy to combined folder with unique name
            combined_output_path = os.path.join(combined_folder, unique_name)
            shutil.copy(output_path, combined_output_path)
            print(f"Copied to combined folder: {combined_output_path}")

# Main function to process selected speakers
def process_speakers(input_base_folder, output_base_folder, speakers, augmentations, probabilities):
    combined_folder = os.path.join(output_base_folder, 'combined')  # Combined folder for all augmented files
    os.makedirs(combined_folder, exist_ok=True)

    for speaker in speakers:
        input_folder = os.path.join(input_base_folder, speaker)
        output_folder = os.path.join(output_base_folder, speaker)

        if os.path.isdir(input_folder):
            print(f"Processing {speaker}...")
            apply_augmentations(input_folder, output_folder, combined_folder, augmentations, probabilities)
            print(f"Finished processing {speaker}\n")

# Define input and output paths
input_base_folder = '/content/drive/My Drive/Deep_Project/new_data'  # Update with your new_data folder path
output_base_folder = '/content/drive/My Drive/Deep_Project/RoyPlayingWithData'

# Define augmentations
augmentations = {
    "white_noise": add_white_noise,
    "pitch_shift": pitch_shift,
    "time_stretch": custom_time_stretch,
    "vtlp": vtlp,
}

# Define probabilities for each augmentation
probabilities = {
    "white_noise": 0.5,
    "pitch_shift": 0.7,
    "time_stretch": 0.2,
    "vtlp": 0.2,
}

# List of speakers to process
speakers = ['speaker1', 'speaker2', 'speaker3', 'speaker4', 'speaker17', 'speaker18', 'speaker19', 'speaker20']

# Process speakers and generate augmented data
process_speakers(input_base_folder, output_base_folder, speakers, augmentations, probabilities)

print("All speakers have been processed and combined folder has been created!")
```
